```python
# 개수 세기
N = int(input())
seq = list(map(int, input().split()))
v = int(input())
cnt = 0
for s in seq:
    if s == v:
        cnt += 1

print(cnt)
```

```python
# 최소, 최대
N = int(input())
lis = list(map(int, input().split()))

print(min(lis), end=' ')
print(max(lis), end='')
```

```python
# 최댓값
max = 0
for i in range(1,10):
    x = int(input())
    if max < x:
        max = x
        idx = i
print(max)
print(idx)
```

```python
# 과제 안 내신 분?
num = list(range(1,31))
for _ in range(28):
    n = int(input())
    num.remove(n)

for l in num:
    print(l)
```

```python
# 나머지
lis = []
for i in range(10):
    x = int(input())
    lis.append(x%42)

print(len(set(lis)))
```

```python
# 평균
N = int(input())
scores = list(map(int, input().split()))
score_revised = []
for s in scores:
    score_revised.append(s/max(scores) * 100)

print(sum(score_revised) / N)
```

```python
# OX퀴즈
N = int(input())
for i in range(N):
    x = str(input())
    sum=0
    count=0
    for j in range(len(x)):
        if x[j] == 'O':
            count += 1
            sum += count
        else:
            count = 0

    print(sum)
```

```python
# 평균은 넘겠지
N = int(input())
for _ in range(N):
    scores = list(map(int,input().split()))
    mean = sum(scores[1:])/scores[0]
    count=0
    for i in range(1,scores[0]+1):
        if scores[i] > mean:
            count += 1

    percent = (count/scores[0])*100
    print('{0:0.3f}%'.format(percent))
```
