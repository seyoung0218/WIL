```python
# 소수 찾기
N = int(input())
num_list = list(map(int, input().split()))
ans = []
if 1 in num_list:
    num_list.remove(1)
    N -= 1 

for i in range(N):
    for j in range(2,num_list[i]):
        if num_list[i] % j == 0:
            ans.append(num_list[i])

k = set(num_list) - set(ans)
print(len(k))

# 구글링
n = int(input())
numbers = map(int, input().split())
sosu = 0
for num in numbers:
    error = 0
    if num > 1 :
        for i in range(2, num):  # 2부터 n-1까지
            if num % i == 0:
                error += 1  # 2부터 n-1까지 나눈 몫이 0이면 error가 증가
        if error == 0:
            sosu += 1  # error가 없으면 소수.
print(sosu)
```

```python
# 소수구하기
M, N = map(int, input().split())

for num in range(M, N+1):
    if num > 1:
        for i in range(2,int(num**0.5) + 1):
            if num % i == 0:
                break 
        else:
            print(num)
```

```python
# 베르트랑 공준
# 시간초과 코드
from math import sqrt
while True:
    n = int(input())
    if n == 0:
        break
    cnt = 0
    for num in range(n+1, 2*n+1):
        for i in range(2, int(sqrt(num)+1)):
            if num % i == 0:
                break
        else:
            cnt += 1

    print(cnt)

# 구글링
def get_sosu(num):
    if num == 1:
        return False
    for i in range(2, int(num**0.5 + 1)):
        if num % i == 0:
            return False
    return True

sosu = [] 
for n in range(1,123456*2):
    if get_sosu(n):
        sosu.append(n)

while True:
    n = int(input())
    if n == 0 :
        break
    
    cnt = 0
    for s in sosu:
        if n < s <= 2*n:
            cnt += 1
    
    print(cnt)
```
